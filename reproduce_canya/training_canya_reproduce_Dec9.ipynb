```python
%%bash
/usr/local/envs/canyaenv/bin/pip install openpyxl
/usr/local/envs/canyaenv/bin/pip install plotly
PYTHONUNBUFFERED=1 stdbuf -oL -eL /usr/local/envs/canyaenv/bin/python -u - <<'PY'



# import libraries
from sklearn.model_selection import train_test_split
import tensorflow as tf, pandas as pd, numpy as np
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from canya.modeling import create_model, get_canya, get_embedded_seqs

import re
from pathlib import Path

# for visualization
import plotly.graph_objects as go
import plotly.io as pio

base = Path("/content/drive/MyDrive/master_thesis/canya-source/canya-reproduce/splits_NNK_data")
weights_dir = Path("/content/drive/MyDrive/master_thesis/canya-source/canya-reproduce/splits_NNK_data/splits_weights")

valid_aas = re.compile(r"^[ACDEFGHIKLMNPQRSTVWY]+$")  # keep 20 natural amino acids

# process aa_sequence and labels
def load_split(ds, split):
    df = pd.read_excel(base / f"{ds}_{split}.xlsx")
    df = df.dropna(subset=["aa_seq", "seed_bh"]) # remove na
    df["aa_seq"] = df["aa_seq"].astype(str).str.strip().str.upper()
    df = df[df["aa_seq"].str.match(valid_aas)]   # keep 20 natural aa

    seqs = df["aa_seq"].tolist()
    X = get_embedded_seqs(seqs)                  # 原 modeling.py 的函数，内含 str_to_vector/dynamic_padding
    y = df["seed_bh"].astype("float32").values
    return X, y

datasets = ["NNK1", "NNK2", "NNK3"]

# training, validation, test(held-out) sets
for ds in datasets:
    # 读取原来的 train 集
    X_full_train, y_full_train = load_split(ds, "train")

    # 从 train 中切出 val（例如 20%，按标签分层）
    X_train, X_val, y_train, y_val = train_test_split(
        X_full_train, y_full_train,
        test_size=0.2,
        random_state=42,
        stratify=y_full_train
    )

    # held‑out test 保持原样
    X_test, y_test = load_split(ds, "test")

    # 从头训练用 create_model()；如果要在预训练权重上微调，用 get_canya("models/model_weights1.h5")
    # choose create_model(), to train from the begining
    model = create_model()
    best_weights_path = weights_dir / f"{ds}_best.h5"

    callbacks = [
        EarlyStopping(patience=3, restore_best_weights=True, monitor="val_loss"),
        ModelCheckpoint(best_weights_path , save_best_only=True, save_weights_only=True, monitor="val_loss")
    ]

    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=30,
        batch_size=256,
        callbacks=callbacks,
        verbose=1
    )
    model.save_weights(weights_dir / f"{ds}_final.h5")
    print(f"{ds} done. best -> {best_weights_path}")

    # visualization
    # 从history对象中提取数据
    history_dict = history.history
    loss_values = history_dict['loss']
    val_loss_values = history_dict['val_loss']
    epochs_range = range(1, len(loss_values)+1)

    # loss plot
    fig_loss=go.Figure()
    fig_loss.add_trace(go.Scatter(x=list(epochs_range), y=loss_values,
                                  mode='lines+markers', name='training loss',
                                  line=dict(color='#4263eb'),
                                  marker=dict(color='#4263eb')))
    fig_loss.add_trace(go.Scatter(x=list(epochs_range), y=val_loss_values,
                                  mode='lines+markers',name='val_loss',
                                  line=dict(color='#f76707'),
                                  marker=dict(color='#f76707')))

    fig_loss.update_layout(title = f"{ds}: train and validation loss",  # ds 是当前 NNK1/2/3
                          xaxis_title='epoch',
                          yaxis_title='loss',
                          width=700,
                          height=400)

    # save the plots
    fig_loss.write_html(weights_dir / f"{ds}_loss.html")


PY
```
